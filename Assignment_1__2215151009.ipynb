```python
#Question 01:Generate two vectors with 15 random floats each, plot them on the same graph, label the axes, and add a title.
import numpy as np
import matplotlib.pyplot as plt


vector1 = np.random.rand(15)
vector2 = np.random.rand(15)
print("Vector 1: \n", vector1)
print("\n Vector 2: \n", vector2)


x = np.arange(len(vector1))
print("\n Index: \n", x)


plt.plot(x, vector1, marker='o', linestyle='-', label="Vector 1")
plt.plot(x, vector2, marker='s', linestyle='--', label="Vector 2")


plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Line Graph of Two Random Vectors")
plt.legend()
plt.grid(True)


plt.show()
```

```python
#Question 02:Create a 4x4 matrix with random values, visualize it as a heatmap, and label the rows and columns.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

matrix = np.random.rand(4, 4)
print("Matrix:\n", matrix)


row_labels = ['Row 1', 'Row 2', 'Row 3', 'Row 4']
col_labels = ['Col 1', 'Col 2', 'Col 3', 'Col 4']


sns.heatmap(matrix, annot=True, cmap='coolwarm', linewidths=0.5,
            xticklabels=col_labels, yticklabels=row_labels)
plt.title('Heatmap of 4x4 Matrix')
plt.show()
```

```python
#Question 03:Generate two 4x4 matrices, perform addition, subtraction, and multiplication, visualize the results using bar plots, and discuss the effects.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

A = np.random.randint(0, 25, size=(4, 4))
B = np.random.randint(0, 25, size=(4, 4))

print("Matrix A:\n", A)
print("\n Matrix B:\n", B)


matrix_sum = A + B
matrix_diff = A - B
matrix_prod = np.dot(A, B)


fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Addition
axes[0].bar(range(1, 17), matrix_sum.flatten(), color='orange')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')

# Subtraction
axes[1].bar(range(1, 17), matrix_diff.flatten(), color='pink')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')

# Multiplication
axes[2].bar(range(1, 17), matrix_prod.flatten(), color='brown')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```

# **Explanation of the Codes:**
**For Question 01:**

plt.legend() adds a legend to distinguish between the two vectors.
plt.grid(True) adds a grid for better readability.
plt.show() displays the graph.

**For Question 02:**

np.random.rand(4, 4) creates a 4x4 matrix with random numbers.
row_labels and col_labels are used to label the rows and columns.
sns.heatmap() creates the heatmap, annot=True: Shows values in each cell, cmap='coolwarm': Applies a color scheme, linewidths=0.5: Adds grid lines, xticklabels and yticklabels: Assign custom labels.
plt.title() adds a title, and plt.show() displays the heatmap.

**For Question 03:**

Two 4x4 matrices (A and B) are created using np.random.randint(0,25) with values between 0 and 25.
Each matrix is flattened into a 1D array using .flatten() for visualization.
Results of the operations are plotted in a 1x3 grid of subplots using plt.subplots(). 04.plt.tight_layout() adjusts spacing between subplots.
Discussion of Effects:

Addition : Combines corresponding elements of A and B.

Subtraction : Represents the difference between corresponding elements of A and B. It shows how much A deviates from B.

Multiplication : Represents the dot product of A and B. It reflects the combined linear transformation of the two matrices.
